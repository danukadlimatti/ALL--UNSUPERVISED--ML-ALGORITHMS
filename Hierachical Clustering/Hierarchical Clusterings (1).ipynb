```python

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
Univ = pd.read_csv("Universities.csv")
```

```python
Univ
```

```python
# Normalization function (0,1)
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
# Univ.iloc[:,1:]
```

```python
# Normalized data frame (considering the numerical part of data)
df_norm = norm_func(Univ.iloc[:,1:])
```

```python
df_norm
```

```python
# import hierarchical clustering libraries
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
dendrogram = sch.dendrogram(sch.linkage(df_norm, method='single'))
```

```python
dendrogram = sch.dendrogram(sch.linkage(df_norm, method='complete'))
```

```python
# create dendrogram
dendrogram = sch.dendrogram(sch.linkage(df_norm, method='average'))
```

```python
# create dendrogram
dendrogram = sch.dendrogram(sch.linkage(df_norm, method='centroid'))
```

```python
# create clusters
hc = AgglomerativeClustering(n_clusters=6, metric = 'euclidean',linkage='complete')
```

```python
hc
```

```python
# df_norm
```

```python
# save clusters for chart
y_hc = hc.fit_predict(df_norm)
```

```python
y_hc
```

```python
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
```

```python
Clusters
```

```python
df_norm['h_clusterid'] = Clusters
df_norm
```

```python
df_norm.sort_values("h_clusterid")
```

```python
df_norm.h_clusterid.value_counts()
```

```python
Univ["Cluster"] = Clusters
```
